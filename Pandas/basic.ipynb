```python
import pandas as pd

# creating a DataFrame (like an Excel Sheet)
data = {
    'Name': ['Arman','John','Alice','Bob'],
    'Age': [23,26,30,35],
    'City': ['Lucknow','New York', 'Paris','Mumbai']
}

df = pd.DataFrame(data)

print(df)
```

```python
# Selecting Data
df['Name']
df[['Name','City']]
df.iloc[0:3]
```

```python
# Modify Data
df['Age'] = df['Age']+1
df['Country'] = 'India'
df.rename(columns={'Name': 'FullName'},inplace=True)
df.head()
```

```python
# Handing Missing Data
df.dropna()
df.fillna(0)
```

```python
# Aggregation
df['Age'].mean()
df['Age'].max()
df['Age'].min()
df['Age'].sum()
df['Age'].count()
df['Age'].describe()
df['Age'].std()
```
